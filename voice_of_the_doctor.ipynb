```python
# Step 2: Use TTS model for text-to-speech output
# This script automates playback of the saved audio file after generation.

import platform
from elevenlabs import save
from pydub import AudioSegment
import tempfile

# Windows audio playback
import winsound  


def play_audio(output_filepath):
    os_name = platform.system()
    try:
        if os_name == "Darwin":  # macOS
            os.system(f'afplay "{output_filepath}"')
        elif os_name == "Windows":
            # Convert MP3 to WAV for Windows playback
            sound = AudioSegment.from_file(output_filepath, format="mp3")
            with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as temp_wav:
                wav_path = temp_wav.name
                sound.export(wav_path, format="wav")
            winsound.PlaySound(wav_path, winsound.SND_FILENAME)
        else:
            raise OSError("Unsupported OS")
    except Exception as e:
        print(f"Error playing audio: {e}")

#{espicially for windows first convert mp3 to wav as mp3 is not supported bt winsound on windows}

def text_to_speech_with_gtts(input_text, output_filepath):
    tts = gTTS(text=input_text, lang="en", slow=False)
    tts.save(output_filepath)
    play_audio(output_filepath)

def text_to_speech_with_elevenlabs(input_text, output_filepath):
    client = ElevenLabs(api_key=ELEVENLABS_API_KEY)
    audio = client.generate(
        text=input_text,
        voice="Aria",
        output_format="mp3_22050_32",
        model="eleven_turbo_v2"
    )
    save(audio, output_filepath)
    play_audio(output_filepath)

# Example usage
input_text = "Hi! This is an automated voice test using AI."
text_to_speech_with_gtts(input_text, "gtts_test.mp3")
text_to_speech_with_elevenlabs(input_text, "elevenlabs_test.mp3")

```
